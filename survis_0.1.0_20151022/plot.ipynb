```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Given data
papers = [
    [2016, 'Clustering techniques'],
    [2015, 'Clustering techniques'],
    [2019, 'Passenger number'],
    [2015, 'Passenger number'],
    [2015, 'Passenger flow'],
    [2018, 'Passenger flow'],
    [2018, 'Travel time'],
    [2015, 'Travel time'],
    [2021, 'Destination prediction'],
    [2018, 'Route mapping']
]

# Convert the data into a pandas DataFrame
df = pd.DataFrame(papers, columns=['Year', 'Paper Type'])

# Define a custom color palette using color names
custom_palette = sns.color_palette(["yellow",  "orange","blue", "red", "green", "gray"])

# Creating a histogram using seaborn with the custom color palette
sns.histplot(data=df, x='Year', hue='Paper Type', palette=custom_palette, multiple='stack')

# Setting x-axis and y-axis labels
plt.xlabel('Year')
plt.ylabel('Number of Papers')

# Displaying the plot
plt.show()
```
